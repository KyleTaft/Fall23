```python
import requests
from bs4 import BeautifulSoup
import os

# Define the base URL of the webpage with audio links

# Create a folder to store the downloaded audio files
output_folder = 'audio_files'

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Loop through each page 
for page_number in range(1, 500):
    # Construct the URL for the current page
    url = f'https://avocet.integrativebiology.natsci.msu.edu/recordings/{page_number}/download?filetype=wav'
    print(f'Downloading page: {url}')
    
    # Send an HTTP GET request to the webpage
    response = requests.get(url)

    # Check if the request was successful (HTTP status code 200)
    if response.status_code == 200:
        # Extract the audio file name from the URL
        audio_name = str(page_number) + "." + url.split('=')[-1]
        print(audio_name)
        audio_path = os.path.join(output_folder, audio_name)  # Create a path to save the audio file

        # Save the audio file to the output folder
        with open(audio_path, 'wb') as audio_file:
            audio_file.write(response.content)
            print(f'Downloaded: {audio_name}')
    else:
        # If the request was not successful, print an error message
        print(f'Failed to download audio from: {url}')

# Print a completion message when all downloads are done
print('Download complete!')
```
