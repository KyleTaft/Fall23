```python
import numpy as np
import matplotlib.pyplot as plt

# Our (0,0,0) and (2,0,3)
x = np.linspace(0, 2, 10)
z = np.linspace(0, 3, 10)
X, Z = np.meshgrid(x, z)

# Calculate v(x,y,z)
Y = np.zeros_like(X) 
U = 3*X*Z 
V = 5*X    
W = 2*Y   


fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=0.3, normalize=True, color='red')
ax.plot_surface(X, Y, Z, alpha=0.5, color='blue')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_xlim(0, 2)
ax.set_ylim(-1, 1) 
ax.set_zlim(0, 3)

plt.show()
```
